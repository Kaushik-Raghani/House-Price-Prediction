```python
Housing_Price=Housing_Price.drop(['Id','PoolQC','MiscFeature','Alley','Fence','MasVnrType','FireplaceQu'],axis=1)
```

```python
c_columns=Housing_Price.select_dtypes(include=['object','category']).columns
print("categorical columns: ")
print(c_columns)
```

```python
label_encoder=LabelEncoder()
```

```python
c_columns=['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
       'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating',
       'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual', 'Functional',
       'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive',
       'SaleType', 'SaleCondition']
for col in c_columns:
    Housing_Price[col]=label_encoder.fit_transform( Housing_Price[col])

Housing_Price
    
```

```python
Housing_Price=Housing_Price.drop(columns=['GarageCars','GarageYrBlt','TotRmsAbvGrd','Foundation','2ndFlrSF','HalfBath','TotalBsmtSF'])
Housing_Price
```

```python
Housing_Price.isnull().sum()
```

```python
Housing_Price['LotFrontage']=Housing_Price['LotFrontage'].fillna(Housing_Price['LotFrontage'].mean())
Housing_Price.isnull().sum()
```

```python
x=Housing_Price.drop(['SalePrice'],axis=1)
y=Housing_Price['SalePrice']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=XGBRegressor(
    objective='reg:squarederror',
    learning_rate=0.01,
    max_depth=6,
    n_estimater=1000,
    subsample=0.65,
    random_state=42
)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error: ",mae)
r2_score=r2_score(y_test,y_pred)
print("R2 Score: ",r2_score)
mse=mean_squared_error(y_test,y_pred)
print("Mean Squared Error: ",mse)
```
